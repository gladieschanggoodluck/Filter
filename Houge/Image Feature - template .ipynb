# Template matching
- cv2.matchTemplate()
- method = CV_TM_CCORR_NORMED

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline


image_color = cv2.imread('meatball.jpg')
cv2.imshow('Find meatball template matching', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width  = image_color.shape[: 2]
```

```python
width
```

```python
plt.imshow(image_color)
```

# Convert to Greyscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
# load template image
template = cv2.imread('meatball_template.jpg', 0)
cv2.imshow("Meatball template", template)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
template.shape
```

```python
w, h = template.shape[ : : -1]
w
```

```python
h
```

# Perform tempalte matching

```python
result = cv2.matchTemplate(image_gray, template, cv2.TM_CCOEFF_NORMED)
result.shape
```

```python
cv2.imshow("Matching Results", result)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Find the indices of the ovject to be find
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
max_loc
```

```python
# Draw a boxarront the object(if found)
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(image_color, top_left, bottom_right, (0, 0, 255), 5)
```

```python
cv2.imshow('meatball located', image_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
